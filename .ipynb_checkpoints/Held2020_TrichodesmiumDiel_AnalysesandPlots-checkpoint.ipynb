```python
# generate the network edges (proteins are the nodes)

proteinsforlineplots = KOoutput.transpose()
proteinsforlineplots.rename(columns=proteinsforlineplots.iloc[0],inplace=True)
dft = proteinsforlineplots.drop('trichotimeofday').reset_index(0)
dft = dft.iloc[:40]
dft['tod'] = dft['index'].str.split('l').apply(lambda x: x[1][:-2]).astype(float)
dft['rep'] = dft['index'].str.split('_').apply(lambda x: x[1]).astype(float)
dft.sort_values(by=["tod","rep"], inplace=True, ascending=True)
dft.reset_index(inplace=True)

sdictneg = {}
sdictall = {}
G_symmetricneg = nx.Graph()
G_symmetricall = nx.Graph()
for name, entry in protDict.items():
    for name2, entry2 in protDict.items():
        s,p = stats.spearmanr(dft[entry], dft[entry2])
        if s < -0.8 and p < 0.05:
            sdictneg[name, name2] = s
            sdictall[name, name2] = s
            G_symmetricneg.add_edge(name, name2)
            G_symmetricall.add_edge(name, name2)
sdictpos = {}

G_symmetricpos = nx.Graph()
for name, entry in protDict.items():
    for name2, entry2 in protDict.items():
        s,p = stats.spearmanr(dft[entry], dft[entry2])
        if s > 0.8 and p < 0.05:
            if name != name2:
                sdictpos[name, name2] = s
                sdictall[name, name2] = s
                G_symmetricpos.add_edge(name, name2)
                G_symmetricall.add_edge(name, name2)   
```

```python
# Number of times a node is present in the shortest path between 2 other nodes - positive correlation        
betCent = nx.closeness_centrality(G_symmetricpos)         

node_color = [100.0 * G_symmetricpos.degree(v) for v in G_symmetricpos]
node_size =  [v * 1500 for v in betCent.values()]
    
pos = nx.spring_layout(G_symmetricpos)    
plt.figure(figsize=(9,9))
nx.draw_networkx(G_symmetricpos, pos=pos, with_labels=True,
                 node_color=node_color,
                 node_size=node_size,
                  font_size = 13,
                cmap='cool')
plt.axis('off')
```

```python
# Number of times a node is present in the shortest path between 2 other nodes  - negative correlations          
betCent = nx.closeness_centrality(G_symmetricneg)         

node_color = [100.0 * G_symmetricneg.degree(v) for v in G_symmetricneg]
node_size =  [v * 1500 for v in betCent.values()]
    
pos = nx.spring_layout(G_symmetricneg)    
plt.figure(figsize=(8,8))
G = nx.draw_networkx(G_symmetricneg, pos=pos, with_labels=True,
                 node_color=node_color,
                 node_size=node_size,
                 font_size = 13,
                cmap = 'cool')
plt.axis('off')
```

## Figure 3. Scatter plots and curve fits, to emphasize the positive and negative correlation relationships

```python

# Defining functions for plots - linear and non-linear fits

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 16}

plt.rc('font', **font)

def plotandoptimize(xs,ys):
    
    dfz = dft.sort_values(by=protDict[xs])
    xdata = dfz[protDict[xs]]
    ydata = dfz[protDict[ys]]
    
    fig, axes = plt.subplots(figsize=(4,4))
    sns.scatterplot(xdata,ydata)

    plt.xlabel(xs)
    plt.ylabel(ys)
    plt.tight_layout()

    # fit an exponential curve

    def func(x, a, b, c):
        #return a * np.log(x) + c
        return a * np.exp(-b * x) + c

    popt, pcov = curve_fit(func, xdata, ydata)
    residuals = ydata- func(xdata, *popt)
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((ydata-np.mean(ydata))**2)
    r_squared = 1 - (ss_res / ss_tot)

    plt.plot(xdata, func(xdata, *popt), 'r-', ms=3, label='a=%5.3f,\nb=%5.3f,\nc=%5.3f' % tuple(popt))

    print(r_squared)
    plt.legend(fontsize=14)        
    
def plotandoptimizelinear(xs,ys):
    
    xdata = dft[protDict[xs]]
    ydata = dft[protDict[ys]]
   
    fig, axes = plt.subplots(figsize=(4,4))
    sns.scatterplot(xdata,ydata)

    plt.xlabel(xs)
    plt.ylabel(ys)
    plt.tight_layout()

    # fit an exponential curve

    def func(x, a, b):
        #return a * np.log(x) + c
        return a*x + b

    popt, pcov = curve_fit(func, xdata, ydata)
    residuals = ydata- func(xdata, *popt)
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((ydata-np.mean(ydata))**2)
    r_squared = 1 - (ss_res / ss_tot)
    print(r_squared)

    plt.plot(xdata, func(xdata, *popt), 'r-', ms=3, label='a=%5.3f,\nb=%5.3f' % tuple(popt))
    plt.legend(fontsize=14)      
                                
```

```python
# then can plot relationships between individual proteins simply by calling the desired function

plotandoptimizelinear('NifH','NifD')
```
